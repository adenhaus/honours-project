```python
def get_full_language_name(input_string):
    language_dict = {
        "en": "English",
        "ar": "Arabic",
        "sw": "Swahili",
        "yo": "Yoruba",
        "ig": "Igbo",
        "pt": "Portuguese",
        "fr": "French",
        "ru": "Russian",
        "ha": "Hausa"
    }

    for language_code in language_dict:
        if "-" + language_code + "-" in input_string:
            return language_dict[language_code]

    return ""
```

```python
import pandas as pd

df = pd.read_csv('train_eng_blueprints_exploded.csv', delimiter='\t')

# --- For translated blurptints:

# df['blueprint_target'] = "Blueprint: " + df['blueprint'].astype(str) + " Verbalisation: " + df['table_text'].astype(str) # use for train
# df['blueprint_target'] = "Blueprint: " + df['blueprint'].astype(str) + " Verbalisation: " + df['target'].astype(str) # use for dev

# --- For english blurptints:

def format_blueprint(row):
    full_language_name = get_full_language_name(row['example_id'])
    return f"Blueprint: {row['blueprint']} Verbalisation ({full_language_name}): {row['table_text']}" # use for train
    # return f"Blueprint: {row['blueprint']} Verbalisation ({full_language_name}): {row['target']})" # use for dev

df['blueprint_target'] = df.apply(format_blueprint, axis=1)

# ---

df.to_csv('train_eng_blueprints_final.csv', sep='\t', index=False)
```
