```python
import pandas as pd
from transformers import AutoTokenizer

model_name = 'google/mt5-small'
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(text):
    return tokenizer(text)


def calc_length_stats(csv_file):
  # Read CSV file
  df = pd.read_csv(csv_file, delimiter='\t')
  df['blueprint_target'] = df['blueprint_target'].astype(str)
  df['linearized_input'] = df['linearized_input'].astype(str)

  # Function to count the elements in 'input_ids'
  def count_elements(tokenized_text):
      return len(tokenized_text['input_ids'])

  # Apply the tokenize function to the 'linearized_input' and 'target' columns
  df['linearized_input'] = df['linearized_input'].apply(tokenize)
  df['blueprint_target'] = df['blueprint_target'].apply(tokenize)

  # Calculate the number of elements in 'input_ids' for each row
  df['linearized_input_length'] = df['linearized_input'].apply(count_elements)
  df['target_length'] = df['blueprint_target'].apply(count_elements)

  # Calculate the average, minimum, and maximum number of elements in 'input_ids' for both columns
  linearized_input_avg_length = df['linearized_input_length'].mean()
  linearized_input_median_length = df['linearized_input_length'].median()
  linearized_input_min_length = df['linearized_input_length'].min()
  linearized_input_max_length = df['linearized_input_length'].max()

  target_avg_length = df['target_length'].mean()
  target_median_length = df['target_length'].median()
  target_min_length = df['target_length'].min()
  target_max_length = df['target_length'].max()

  # Print the results
  print("Linearized Input Column:")
  print(f"Average Input_ids Length: {linearized_input_avg_length:.2f}")
  print(f"Median Input_ids Length: {linearized_input_median_length:}")
  print(f"Minimum Input_ids Length: {linearized_input_min_length}")
  print(f"Maximum Input_ids Length: {linearized_input_max_length}\n")

  print("Target Column:")
  print(f"Average Input_ids Length: {target_avg_length:.2f}")
  print(f"Median Input_ids Length: {target_median_length:}")
  print(f"Minimum Input_ids Length: {target_min_length}")
  print(f"Maximum Input_ids Length: {target_max_length}")
```
