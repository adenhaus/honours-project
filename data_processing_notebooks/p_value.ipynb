```python
import pandas as pd
from scipy.stats import pearsonr

# Read the data from the text file into a DataFrame
data = pd.read_csv('stata.txt', delimiter=':', skipinitialspace=True, names=['key', 'value'])

# Extract predictions and references from DataFrame
predictions = data[data['key'] == 'prediction']['value'].astype(float)
references = data[data['key'] == 'reference']['value'].astype(float)

# Calculate Pearson correlation coefficient and its p-value
pearson_corr, p_value = pearsonr(predictions, references)

# Check if correlation is significant at p < 0.05
if p_value < 0.01:
    significance = "significant"
else:
    significance = "not significant"

print("Pearson Correlation Coefficient:", pearson_corr)
print("P-value:", p_value)
print("The correlation is", significance, "at the p < 0.05 level.")
```
