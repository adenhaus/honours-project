```python
from transformers import MT5ForSequenceClassification, MT5Tokenizer

# Define the model name
model_name = "google/mt5-large"

# Load the pre-trained model and tokenizer
model = MT5ForSequenceClassification.from_pretrained(model_name)
tokenizer = MT5Tokenizer.from_pretrained(model_name)

# Specify the directory to save the model and tokenizer
save_directory = "./saved_mt5_model"

# Save the model and tokenizer to the specified directory
model.save_pretrained(save_directory)
tokenizer.save_pretrained(save_directory)

print(f"Model and tokenizer saved to {save_directory}")
```
