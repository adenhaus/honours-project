```python
import pandas as pd

def process_csv(input_file, output_file):
    # Read CSV file
    df = pd.read_json(input_file)

    # Drop rows where "-en-" is not present in the 'example_id' column
    df = df[df['example_id'].str.contains('-en-')]

    if input_file == 'train_skip_no_refs_exploded.json':
      df = df.rename(columns={'table_text': 'target'}, inplace=False)
      df = df.drop(columns=['example_id', 'title', 'unit_of_measure', 'chart_type', 'was_translated', 'table_data'])
    else:
      df = df.drop(columns=['example_id', 'title', 'unit_of_measure', 'chart_type', 'was_translated', 'table_data', 'table_text'])

    # Write the filtered DataFrame to a new CSV file
    df.to_csv(output_file, index=False, sep='\t')

process_csv('dev_with_targets.json', 'eng_dev.csv')
process_csv('train_skip_no_refs_exploded.json', 'eng_train.csv')
```
