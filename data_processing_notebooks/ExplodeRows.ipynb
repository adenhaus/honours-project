```python
import pandas as pd
import ast

df = pd.read_csv('train_skip_no_refs_eng_blueprints_all.csv', delimiter='\t')

df['table_text'] = df['table_text'].apply(ast.literal_eval)
df['blueprint'] = df['blueprint'].apply(ast.literal_eval)

mask = df.apply(lambda row: len(row['blueprint']) == len(row['table_text']), axis=1)

# Use the mask to filter rows
df = df[mask]

df_exploded = df.explode(['blueprint', 'table_text'])

print(len(df_exploded))

df_exploded.to_csv('blueprints_exploded.csv', sep="\t", index=False)
```
