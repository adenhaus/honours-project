```python
import pandas as pd

# Read CSV file
csv_file_path = 'test_full.csv'
df = pd.read_csv(csv_file_path, sep='\t')

# List of strings to filter on in the example_id column
languages = ["en", "yo", "ig", "sw", "ha", "fr", "ar", "pt"]

# Create separate DataFrames for each language and write to CSV
for lang in languages:
    lang_df = df[df['example_id'].str.contains(lang)]

    # Write the language-specific DataFrame to a new CSV file
    lang_csv_file_path = f'{lang}_test.csv'
    lang_df.to_csv(lang_csv_file_path, sep='\t', index=False)

    print(f'DataFrame for {lang} written to: {lang_csv_file_path}')
```
